```python
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()

X = dataset.data
y = dataset.target

feature_names = dataset.feature_names
target_names = dataset.target_names

print(dataset.DESCR)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)

rfc = RandomForestClassifier(random_state=42)
rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print(f"Accuracy: {acc * 100:.2f}%")
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,4))

plt.scatter(range(len(y_train)), y_train, color="blue", label="true labels", alpha=0.5, marker='o')
plt.scatter([x + 0.1 for x in range(len(y_train))], rfc.predict(X_train), color="pink", label="Predicted Labels", alpha=0.6, marker='x')

plt.title("Breast Cancer Predictions (Training Set)")
plt.xlabel("Sample Index")
plt.ylabel("Labels")
plt.legend()
plt.show()
```
